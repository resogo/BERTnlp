```python
path = ".\deid_surrogate_test_all_groundtruth_version2\deid_surrogate_test_all_groundtruth_version2.xml"
file = open(path, "r")
contents = file.read()
soup = BeautifulSoup(contents, "xml")
phis = soup.find_all('PHI')
len(phis)
```

```python
import re
removeTags = re.sub('<PHI TYPE=".+">','',contents)
removeTags = re.sub('<.+>','',removeTags)
print(removeTags)
```

# Format Data

```python
values = re.findall('[^a-z0-9\n]+ :', removeTags) #this forms the lable set
content = re.split('[^a-z0-9\n]+ :', removeTags) #this is the corresponding sentences for each able 
content.pop(0) #remove first element wwhich doesn't correspond to any label
for i, val in enumerate(values):
    if len(val)<6:
        values.remove(val)
        content.pop(i)
print(values)
len(values)
```

```python
#This finds the number of unique lables which we will use as our lables
matches = [values[0]]
for x in values:
    if x not in matches:
        matches.append(x)
```

## Prepare data in arrays to convert to csv

```python
labels = []
texts = []
alist = []
indexlist = []
id = 1
for index, entry in enumerate(values):
    if len(entry)>4:
        current_label = matches.index(entry)
        if current_label!=83 and current_label!=36:
            s = content[index+1]
            s = s.replace('\n', '')
            if len(s)>1:
                labels.append(current_label)
                texts.append(s)
                alist.append('a')
                indexlist.append(id)
                id = id+1
    
```

```python
#split data between train and dev files
import math
l = len(indexlist)
print(l)
l1 = math.floor(l*9/10)
print(l1)
l2 = l-l1
#l3 = math.floor(l2/2)+l1;
```

```python
#make train.tsv
import pandas
df = pandas.DataFrame(data={"col1": indexlist[0:-l2], "col2": labels[0:-l2], "col3": alist[0:-l2], "col4": texts[0:-l2]})
df.to_csv("./train.csv", sep=',',index=False)

df = pandas.read_csv('./train.csv')
df = df.iloc[1:]
df.to_csv('./train.tsv', sep='\t', index=False, header=False)
```

```python
#make dev.tsv
import pandas
df = pandas.DataFrame(data={"col1": indexlist[l2+1:l3], "col2": labels[l2+1:l3], "col3": alist[l2+1:l3], "col4": texts[l2+1:l3]})
df.to_csv("./dev.csv", sep=',',index=False)

df = pandas.read_csv('./dev.csv')
df.to_csv('./dev.tsv', sep='\t', index=False, header=False)
```
