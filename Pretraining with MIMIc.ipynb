```python
import re
def tokenize_note(sent):
    tokenizer = re.compile('\w+|\*\*|[^\s\w]')
    tokens = tokenizer.findall(sent)
    return tokens
```

```python
count = 0
with open("./pretraining_files/mimictrain.txt", 'r') as myfile:
    for line in myfile:
        count+=len(tokenize_note(line))
```
